```python
import math
sequence=[
    ["F", "S", "S" ,"F", "F" ,"S" ,"S" ,"S", "F", "S"],
    ["F", "F", "S", "F", "F", "F", "S", "S", "F", "F"],
    ["F", "S", "S", "S", "F", "S", "S", "F", "S", "S"],
    ["S", "F", "F", "F", "S", "F", "F", "S", "F", "S"],
    ["S", "S", "S", "F", "S", "F", "S", "S", "F", "S"]    
]
Theta_1=0.55
Theta_2=0.7
#------------For sequence 1--------------------
n=len(sequence[0])
success=0
failure=0
archerA_Success=0
archerB_Success=0
archerA_Failure=0
archerB_Failure=0
for i in range(len(sequence)):
    success=sequence[i].count("S")
    failure=sequence[i].count("F")
    print("For Sequence :",i+1)
    #for archer A
    PBerA=(math.comb(n,success))*(Theta_1)**success * (1-Theta_1)**failure
    #for archer B
    PBerB=(math.comb(n,success))*(Theta_2)**success * (1-Theta_2)**failure
    print("PBernoulli_A: ",PBerA)
    print("PBernoulli_B: ",PBerB)
    print("The probability that the first sequence is generated from player A: ",(PBerA/(PBerA+PBerB)))
    print("The probability that the first sequence is generated from player B: ",(PBerB/(PBerA+PBerB)))
    print("The frequency of success for Archer A: ",success*(PBerA/(PBerA+PBerB)))
    print("The frequency of failure for Archer A: ",failure*(PBerA/(PBerA+PBerB)))
    print("The frequency of success for Archer B: ",success*(PBerB/(PBerA+PBerB)))
    print("The frequency of failure for Archer B: ",failure*(PBerB/(PBerA+PBerB)))
    if((PBerA/(PBerA+PBerB))>(PBerB/(PBerA+PBerB))):
        print("Sequence ",i+1," is most likely to be shot by A")
    else:
        print("Sequence ",i+1," is most likely to be shot by B")
    archerA_Success+=success*(PBerA/(PBerA+PBerB))
    archerB_Success+=success*(PBerB/(PBerA+PBerB))
    archerA_Failure+=failure*(PBerA/(PBerA+PBerB))
    archerB_Failure+=failure*(PBerB/(PBerA+PBerB))
print("archerA_Success:",archerA_Success)
print("archerB_Success:",archerB_Success)
print("archerA_Failure:",archerA_Failure)
print("archerB_Failure:",archerB_Failure)
print("Theta_1(1)",(archerA_Success)/(archerA_Success+archerA_Failure)) 
print("Theta_2(1)",(archerB_Success)/(archerB_Success+archerB_Failure))
```
