```python
#Tau-B----------------------------------------------
import math
sample_1_toSort=[30,40,40,20,12,31,10]
sample_2=[160,160,170,150,110,160,100]
sample_unsorted=[30,40,40,20,12,31,10]
sorted_3=[]
sample_1_toSort.sort()
repeated_dic={}
for i in range(len(sample_1_toSort)):
    index=sample_unsorted.index(sample_1_toSort[i])
    sorted_3.insert(i,sample_2[index])
print(sorted_3)
print(sample_1_toSort)

#-Concordant 
concordant_count=0
concordant_pair=[]
for i in range(len(sorted_3)):
    for j in range(len(sorted_3)-1):
        if(sorted_3[j+1]>sorted_3[i] and j>=i):
            concordant_pair.append([sorted_3[i],sorted_3[j+1]])
            concordant_count+=1
print("concordant_pair:",concordant_pair)
print("concordant_count: ",concordant_count)
#Discordant Pairs
discordant_count=0
discordant_pair=[]
for i in range(len(sorted_3)):
    for j in range(len(sorted_3)-1):
        if(sorted_3[j+1]<sorted_3[i] and j>=i):
            discordant_pair.append([sorted_3[i],sorted_3[j+1]])
            discordant_count+=1
print("discordant_pair: ",discordant_pair)
print("discordant_count: ",discordant_count)


#Kendall Tau A
n=len(sample_2)
tau=(concordant_count-discordant_count)/(0.5*n*(n-1))
print("Tau-A: ",tau)

sample_A_Pair=0
sample_B_Pair=0
for i in range(len(sample_1_toSort)):
     for j in range(len(sample_1_toSort)-1):
        if(sample_1_toSort[i]!=sample_1_toSort[j+1] and j>=i):
            sample_A_Pair+=1    
print("sample_A_Pair:",sample_A_Pair)

for i in range(len(sample_2)):
     for j in range(len(sample_2)-1):
        if(sample_2[i]!=sample_2[j+1] and j>=i):
            sample_B_Pair+=1    
print("sample_B_Pair:",sample_B_Pair)
taub=(concordant_count-discordant_count)/(math.sqrt(sample_A_Pair)*math.sqrt(sample_B_Pair))
print("TauB: ",taub)
```
